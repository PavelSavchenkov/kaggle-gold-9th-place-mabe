```python
from collections import defaultdict
from pathlib import Path

import numpy as np
from tqdm import tqdm

from common.config_utils import base_model_from_file
from common.constants import REMOVE_25_FPS_ADAP_SNAIL_SUFFIX
from common.ensemble_building_primitives import EnsembleObjective
from common.metrics_common import OOF_Metrics
from dl.ensembling_dl import EnsembleApproachDL
from dl.oof_utils import OOF_Full, get_best_oof_per_action_lab
from dl.postprocess import get_train_config_dl, ratio_to_step, step_to_ratio
from dl.submission import Submission, SubmissionModel

SRC = Path("submissions/dl-mixed-ens12-12-dec")
DST = Path("submissions/dl-mixed-ens12-full-7seeds-adap-snail")
SEEDS = 7

submission = base_model_from_file(Submission, SRC / "submission.json")

new_models = []
for model in tqdm(submission.models, desc="Converting to full models..."):
    name = model.name()
    ratios = []
    for f in model.folds():
        step = model.get_step(f)
        config = get_train_config_dl(name=name, cv=f)
        assert config is not None
        ratio = step_to_ratio(config=config, step=step)
        ratios.append(ratio)
    median_ratio = float(np.median(ratios))
    for seed in range(SEEDS):
        dst_name = f"{name}-full-seed{seed}"
        config = get_train_config_dl(name=dst_name, cv=0)
        assert config is not None, f"name={dst_name}"
        step = ratio_to_step(config=config, ratio=median_ratio)
        dst_model = SubmissionModel()
        dst_model.lab = model.lab
        dst_model.action = model.action
        dst_model.coef = model.coef / SEEDS
        th = model.threshold()
        dst_model.oof_full = OOF_Full(
            name=dst_name,
            step_per_fold={0: step},
            metrics=OOF_Metrics(oof_th=th, th_per_fold={0: th}),
        )
        new_models.append(dst_model)

submission.models = new_models
submission.write_dir(dst=DST)
```

```python
import copy
from pathlib import Path
import shutil

from common.config_utils import base_model_from_file
from postprocess.submission_utils import Submission as SubmissionGBDT

SRC = Path("submissions/gbdt-mixed-ens6-12-dec")
DST = Path("submissions/gbdt-mixed-ens6-12-dec-full")
DST.mkdir(exist_ok=True, parents=True)

ref_submission_path = Path("submissions/full-models-26-nov-fix")
# ref_submission = base_model_from_file(
#     SubmissionGBDT, ref_submission_path / "submission.json"
# )

submission = base_model_from_file(SubmissionGBDT, SRC / "submission.json")

submission_full = copy.deepcopy(submission)
submission_full.models.clear()

for model in submission.models:
    copy_from = SRC / model.name
    full_name = f"{model.name}-full"
    full_model_path = ref_submission_path / full_name
    if not full_model_path.exists():
        print(f"Missing {full_model_path}")
    else:
        model.name = full_name
        copy_from = ref_submission_path / model.name
    copy_to = DST / model.name
    if not copy_to.exists():
        shutil.copytree(copy_from, DST / model.name)
    submission_full.models.append(model)
submission_full.write_dir(DST)
```
